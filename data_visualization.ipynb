```python
import json
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

with open("data_raw/News_Dataset_Vietnamese.json", "r", encoding="utf-8") as f:
    data = json.load(f)

df = pd.DataFrame(data)

print("Columns:", df.columns)

# --- Tính độ dài bài báo ---
df['article_length'] = df['Contents'].astype(str).apply(lambda x: len(x.split()))

# --- Vẽ histogram ---
plt.figure(figsize=(10,6))
sns.histplot(df['article_length'], bins=60, kde=True)
plt.title("Phân bố độ dài bài báo News Dataset Vietnamese (Article Length Distribution)")
plt.xlabel("Số từ trong bài báo")
plt.ylabel("Tần suất")
plt.show()
```
