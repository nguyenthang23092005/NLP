```python
import json
import matplotlib.pyplot as plt

with open("models/mt5-lora-full/checkpoint-7728/trainer_state.json", "r", encoding="utf-8") as f:
    data = json.load(f)
log = data["log_history"]
```

```python
steps = [x["step"] for x in log if "loss" in x]
loss = [x["loss"] for x in log if "loss" in x]

plt.figure(figsize=(8,5))
plt.plot(steps, loss)
plt.xlabel("Step")
plt.ylabel("Training Loss")
plt.title("Training Loss Curve")
plt.grid()
plt.show()
```

```python
eval_steps = [x["step"] for x in log if "eval_loss" in x]
eval_loss = [x["eval_loss"] for x in log if "eval_loss" in x]

plt.figure(figsize=(8,5))
plt.plot(eval_steps, eval_loss)
plt.xlabel("Step")
plt.ylabel("Evaluation Loss")
plt.title("Evaluation Loss Curve")
plt.grid()
plt.show()
```

```python
eval_steps = [x["step"] for x in log if "eval_rouge1" in x]
eval_r1 = [x["eval_rouge1"] for x in log if "eval_rouge1" in x]

plt.figure(figsize=(8,5))
plt.plot(eval_steps, eval_r1)
plt.xlabel("Step")
plt.ylabel("ROUGE-1")
plt.title("Evaluation ROUGE-1 Curve")
plt.grid()
plt.show()
```

```python
eval_steps = [x["step"] for x in log if "eval_rouge2" in x]
eval_r2 = [x["eval_rouge2"] for x in log if "eval_rouge2" in x]

plt.figure(figsize=(8,5))
plt.plot(eval_steps, eval_r2)
plt.xlabel("Step")
plt.ylabel("ROUGE-2")
plt.title("Evaluation ROUGE-2 Curve")
plt.grid()
plt.show()
```

```python
eval_steps = [x["step"] for x in log if "eval_rougeL" in x]
eval_rl = [x["eval_rougeL"] for x in log if "eval_rougeL" in x]

plt.figure(figsize=(8,5))
plt.plot(eval_steps, eval_rl)
plt.xlabel("Step")
plt.ylabel("ROUGE-L")
plt.title("ROUGE-L Curve")
plt.grid()
plt.show()
```

```python
lr_steps = [x["step"] for x in log if "learning_rate" in x]
lr_values = [x["learning_rate"] for x in log if "learning_rate" in x]

plt.figure(figsize=(8,5))
plt.plot(lr_steps, lr_values)
plt.xlabel("Step")
plt.ylabel("Learning Rate")
plt.title("Learning Rate Schedule")
plt.grid()
plt.show()
```

```python
steps = [x["step"] for x in log if "grad_norm" in x]
grad_norm = [x["grad_norm"] for x in log if "grad_norm" in x]

plt.figure(figsize=(8,5))
plt.plot(steps, grad_norm)
plt.xlabel("Step")
plt.ylabel("Gradient Norm")
plt.title("Gradient Norm Curve")
plt.grid()
plt.show()
```
